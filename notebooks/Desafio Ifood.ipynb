## Temos dados nulos nessa base? Será que eles indicam algo? O que fazer com eles?

```python
# Verificando se temos dados nulos
df.isna().sum()
```

Podemos ver que algumas colunas possuem valores nulos. As colunas que apresentam dados nulos são:

- marital_Divorced
- marital_Married
- marital_Single
- marital_Together
- marital_Widow
- education_2n Cycle    
- education_Basic
- education_Graduation
- education_Master
- education_PhD

Os nulos acima podem indicar que tal característica não faz parte daquela pessoa. Por exemplo, uma pessoa divorciada  
tem um determinado valor ou um valor nulo

```python
# Contando os valores únicos de uma das colunas nulas. 
df["marital_Divorced"].value_counts()
```

Observando a coluna marital_Divorced, vemos que ela possuí apenas o número 1.0 que aparece 214 vezes, e o resto dos valores são nulos.  
Isso é um indício que o nulo significa que a pessoa não possuí aquela determinada variável.

Podemos substituir o NaN (valores nulos) pelos valores 0 para faltante e 1 para presente.

```python
colunas_com_nulos = [
    "marital_Divorced",
    "marital_Married",
    "marital_Single",
    "marital_Together",
    "marital_Widow",
    "education_2n Cycle",
    "education_Basic",
    "education_Graduation",
    "education_Master",
    "education_PhD"
]

for coluna in colunas_com_nulos:
    df[coluna] = np.where(pd.isna(df[coluna]), 0, 1)
```
