Os nulos acima podem indicar que tal característica não faz parte daquela pessoa. Por exemplo, uma pessoa divorciada  
tem um determinado valor ou um valor nulo

```python
# Contando os valores únicos de uma das colunas nulas. 
df["marital_Divorced"].value_counts()
```

Observando a coluna marital_Divorced, vemos que ela possuí apenas o número 1.0 que aparece 214 vezes, e o resto dos valores são nulos.  
Isso é um indício que o nulo significa que a pessoa não possuí aquela determinada variável.

Podemos substituir o NaN (valores nulos) pelos valores 0 para faltante e 1 para presente.

```python
colunas_com_nulos = [
    "marital_Divorced",
    "marital_Married",
    "marital_Single",
    "marital_Together",
    "marital_Widow",
    "education_2n Cycle",
    "education_Basic",
    "education_Graduation",
    "education_Master",
    "education_PhD"
]

for coluna in colunas_com_nulos:
    df[coluna] = np.where(pd.isna(df[coluna]), 0, 1)
```

## Qual é a distribuição de salário na nossa base? Temos clientes que ganham muito bem e outros que não ganham muito bem?

```python
# Criando o histograma
sns.histplot(x=df["Income"], bins=10);
```

```python
# Criando um boxplot
sns.boxplot(df["Income"]);
```

Além de verificar o histograma e o boxplot, vamos cálcular o coeficiente de variação e o skew. 

O coeficiente de variação é interessante pois não está atrelado a uma unidade de medida então não precisamos ter ideia da grandeza da média para  
entender se o valor é grande ou pequeno. O CV pode ser dado em decimal ou porcentagem.

O skew é a assimetria da distribuição dos dados em relação a sua média, o skew mostra de uma forma numérica onde está a assimetria dos nossos dados.  
Dizemos que os dados são aproximadamente normais (simétricos) se -1 < Skew < 1.

```python
# Calculando o Coeficiente de Variação
# Quanto mais próximo de 1 (100%) mais disperso os dados são
# em relação à média.
df["Income"].std()/df["Income"].mean()
```

```python
# Calculando o skew (assimetria) da distribuição dos dados
df["Income"].skew()
```

Com base nas análises acima, podemos ver que temos uma boa distribuição dos dados, focando no skew podemos ver que temos aproximadamente uma normal.  
Podemos ver também que poucas pessoas ganham pouco e poucas pessoas ganham muito.
